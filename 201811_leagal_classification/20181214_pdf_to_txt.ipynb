# 必要なモジュールの導入

```python
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfpage import PDFPage
from io import StringIO
import os
import re
import mojimoji
```

### pdfの名前一覧を取得

```python
import glob
dir_name = "death"
os.chdir('data/'+dir_name+'_pdf')
name = glob.glob("*")
print(len(name))
os.chdir('..')
os.chdir('..')
```

### 取得した名前を利用してpdf⇒txt

```python
count = 0
for i in range(len(name)):
    input_path = "data/"+dir_name+"_pdf/"+str(name[i])
    rsrcmgr = PDFResourceManager()
    codec = 'z-8'
    params = LAParams()

    # pdf to text
    with StringIO() as output:
        device = TextConverter(rsrcmgr, output, codec=codec, laparams=params)
        with open(input_path, 'rb') as input:
            interpreter = PDFPageInterpreter(rsrcmgr, device)
            for page in PDFPage.get_pages(input):
                interpreter.process_page(page)
                # 確認用
                #print(output.getvalue())
                text = output.getvalue()
        device.close()
    #テキスト成形
    # (罪となるべき事実）を検索
    #index1 = text.find("（罪となるべき事実）")
    # （各犯行に至る経緯及び罪となるべき事実）
    #index2 = text.find("（各犯行に至る経緯及び罪となるべき事実）")
    # （犯罪事実）
    #index3 = text.find("（犯罪事実）")
    # 罪となるべき事実を検索
    #index4 = text.find("罪となるべき事実")
    #if (index1 != -1)or(index2 != -1)or(index3 != -1)or(index4 != -1):
    #    print(count)
    
    # 漢数字を削除
    table = text.maketrans({
        '一': '', #左が置換したい文字、右が新しい文字。
        '二': '', #左が置換したい文字、右が新しい文字。
        '三': '', #左が置換したい文字、右が新しい文字。
        '四': '', #左が置換したい文字、右が新しい文字。
        '五': '', #左が置換したい文字、右が新しい文字。
        '六': '', #左が置換したい文字、右が新しい文字。
        '七': '', #左が置換したい文字、右が新しい文字。
        '八': '', #左が置換したい文字、右が新しい文字。
        '九': '', #左が置換したい文字、右が新しい文字。
        '1': '', #左が置換したい文字、右が新しい文字。
        '0': '', #左が置換したい文字、右が新しい文字。
        '2': '', #左が置換したい文字、右が新しい文字。
        '3': '', #左が置換したい文字、右が新しい文字。
        '4': '', #左が置換したい文字、右が新しい文字。
        '5': '', #左が置換したい文字、右が新しい文字。
        '6': '', #左が置換したい文字、右が新しい文字。
        '7': '', #左が置換したい文字、右が新しい文字。
        '8': '', #左が置換したい文字、右が新しい文字。
        '9': '', #左が置換したい文字、右が新しい文字。
        '０': '', #左が置換したい文字、右が新しい文字。
        '１': '', #左が置換したい文字、右が新しい文字。
        '２': '', #左が置換したい文字、右が新しい文字。
        '３': '', #左が置換したい文字、右が新しい文字。
        '４': '', #左が置換したい文字、右が新しい文字。
        '５': '', #左が置換したい文字、右が新しい文字。
        '６': '', #左が置換したい文字、右が新しい文字。
        '７': '', #左が置換したい文字、右が新しい文字。
        '８': '', #左が置換したい文字、右が新しい文字。
        '９': '', #左が置換したい文字、右が新しい文字。
        '〇': '', #左が置換したい文字、右が新しい文字。
        '、': '', #左が置換したい文字、右が新しい文字。
        '。': '', #左が置換したい文字、右が新しい文字。
        '｡': '', #左が置換したい文字、右が新しい文字。
        '｢': '', #左が置換したい文字、右が新しい文字。
        '｣': '', #左が置換したい文字、右が新しい文字。
        #'ころ': '', #左が置換したい文字、右が新しい文字。
    })
    text = text.translate(table)
    text = re.sub('([あ-んア-ン一-龥ー])\s+((?=[あ-んア-ン一-龥ー]))',r'\1\2', text)
    #print(text)
    # "ころ"を置換
    text = text.replace("ころ","")
    # アルファベットを置換
    text = mojimoji.zen_to_han(text)
    text = re.sub(r'[a-z]+', "", text)
    text = re.sub(r'[A-Z]+', "", text)
    # "証拠の標目"以下を分離
    text_ = text.split("証拠の標目")
    if len(text_)==2:
        #print(text_[0])
        text_ = text_[0].split("理由",1)
        if len(text_)==2:
            None
        else:
            print(len(text_))
            text_ = text_[0].split("理",1)
            print(text_)



        path_w = "data/"+dir_name+"_txt/"+dir_name+"_"+str(i)+".txt"
        with open(path_w, mode='w',encoding="utf-8") as f:
            f.write(text_[1])
            print(path_w)
```
